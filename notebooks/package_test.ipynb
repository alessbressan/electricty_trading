## Getting data from API

```python
from nyisotoolkit import NYISOData

url = 'http://mis.nyiso.com/public/'

df = NYISOData(dataset='asp_dam', year='2019').df
df_da_prices = df['NYCA Regulation Capacity ($/MWHr)',  'NORTH']

df_da_prices.head
```

```python
df = NYISOData(dataset='asp_rt', year='2019').df
df_rt_prices = df['NYCA Regulation Capacity ($/MWHr)',  'NORTH']

df_rt_prices.head()
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2,1, figsize=(15,8))
ax1.plot(df_da_prices, color = 'blue')
ax1.set_title('Day Ahead Prices')

ax2.plot(df_rt_prices, color = 'red')
ax2.set_title('Real Time Prices')

plt.show()
```
